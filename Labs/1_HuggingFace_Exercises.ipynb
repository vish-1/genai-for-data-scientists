# Exercises on Using HuggingFace's Transformers

## Exercise 1: Downloading and Prompting T5
- Objective: Familiarize yourself with the T5 model, download it using HuggingFace's transformers library, and use it for a simple task.
- Tasks:
    1. Install the HuggingFace `transformers` library.
    2. Import the necessary modules for T5.
    3. Download the T5 model using the `transformers` library.
    4. Write a function to prompt T5 with a simple translation task (e.g., translate a sentence from English to French).

## Exercise 2: Transfer Learning with BERT
- Objective: Understand the concept of transfer learning and apply it using the BERT model for a classification task.
- Tasks:
    1. Choose a dataset for a binary classification task (e.g., sentiment analysis).
    2. Load the pre-trained BERT model from the HuggingFace library.
    3. Fine-tune BERT on your chosen dataset for the classification task.
    4. Evaluate the performance of your model on a test set.

## Exercise 3: Distillation of T5-small ( Optional)

- Objective: Learn about model distillation and apply it to the T5-small model to create a smaller, more efficient version.
- Tasks:
    1. Understand the concept of model distillation.
    2. Use the HuggingFace `transformers` and `distillation` libraries to distill BERT.
    3. Compare the performance and size of the distilled model to the original BERT model.

## Exercise 4: Using ROUGE for Evaluation
- Objective: Learn about the ROUGE metric and use it to evaluate a text summarization model.
- Tasks:
    1. Understand what the ROUGE metric is and its importance in NLP.
    2. Choose a text summarization model from HuggingFace's library.
    3. Summarize a set of documents using this model.
    4. Evaluate the summaries using the ROUGE metric.

## Exercise 5: Exploring BLEU for Machine Translation
- Objective: Learn about the BLEU metric and use it to evaluate a machine translation model.
- Tasks:
    1. Understand what the BLEU metric is and its significance in machine translation.
    2. Choose a machine translation model from HuggingFace's library.
    3. Translate a set of sentences from one language to another using this model.
    4. Evaluate the translations using the BLEU metric.
